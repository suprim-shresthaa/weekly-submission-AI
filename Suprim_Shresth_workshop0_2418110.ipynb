```python
#Task 2
def generate_permutation(s):
    result=[]
    def btrack(start):
        output=set()

        for i in range(start,len(s)):
            if s[i] not in output:
                output.add(s[i])
                s[start], s[i] = s[i], s[start]
                btrack(start+1)
                s[start], s[i] = s[i], s[start]

        if start == len(s)-1:
            result.append(''.join(s))
            return #return gare pachi chai recursion rokinxa


    s=list(s)
    btrack(0)
    return result

print(generate_permutation("abc"))
print(generate_permutation("abc"))
```

```python
def calculate_directory_size(directory):
    size = 0
    for a in directory:
        if isinstance(directory[a],dict):
            size += calculate_directory_size(directory[a])
        else:
            size += directory[a]
    return size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

size = calculate_directory_size(directory_structure)
print(f"Total size of directory: {size} KB")

```

```python
#Task 1
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0]=0

    for coin in coins:
        for i in range(coin, amount +1):
             dp[i]=min(dp[i],dp[i-coin]+1)
    return dp[amount] if dp[amount]!=float('inf') else-1
res = min_coins([1,2,5],11)
print(res)


```

```python
# Task 2
def longest_common_subsequence(s1, s2):
    a, b = len(s1), len(s2)

    dp = [[0] * (b+1) for _ in range(a+1)]

    for i in range(1, a+1):
        for j in range(1, b+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1]+1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[a][b]

s1 = "abcde"
s2 = "ace"
res = longest_common_subsequence(s1, s2)
print(res)

```

```python
#Task 3
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0]*(capacity + 1) for _ in range(n+1)]

    for i in range(1,n+1):
        for w in range(1,capacity +1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], values[i - 1] + dp[i - 1][w - weights[i - 1]])
            else:
                dp[i][w] = dp[i - 1][w]
    return dp[n][capacity]

weights=[1,3,4,5]
values=[1,4,5,7]
capacity=7
res = knapsack(weights, values, capacity)
print(res)
```
